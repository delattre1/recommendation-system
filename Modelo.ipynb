```python
#importando as bibilhotecas necessarias
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#importando o banco de dados ultilizado
dados= pd.read_csv("books.csv",header=None)
dados.columns=["bookID","title","authors","average_rating","isbn","isbn13","language_code","#num_pages","rating_counts","text_reviews_count"]
dados.head()
```

```python
#removendo colunas nao importantes para a analise
dados = dados.drop(["bookID","isbn","isbn13","text_reviews_count"],axis=1)
dados.head()
```

```python
#Tratando todos os livros com o mesmo nome como se fossem um so
#antes, garantindo que os livros discartados seram aqueles com o menor rating
dados.sort_values("average_rating",ascending=False, inplace = True)
dados.drop_duplicates(subset="title",keep = False, inplace = True)
dados.head()
```
